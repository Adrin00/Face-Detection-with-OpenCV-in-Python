#cv2.imread()
1. syntax: cv2.imread(file path, flag)
2. file path: denotes location of the image to be read.
3. flag: specifies the way in which to read the image (color- 3 channels(RGB), grayscale- 1 channel(Black and White) etc)
4. return: returns an image that is loaded from the specified location.


#cv2.imshow()
1. syntax: cv2.imshow(windowname, image to be displayed)
2. Window name: name of the window in which to dispaly the file in.
3. return: no return value to program but displays the image on the screen.


##cv2.imwrite()
1. syntax: cv2.imwrite(filename, image to be saved)
2. file name:name of the file that will be saved (Needs file extension).
3. return: returns a Boolean True if the image was saved successfully.

#Face Detection with Python using OpenCV
#coding part

```python
#Import the OpenCV library
import cv2
```

```python
#Using the pre-trained classifier from OpenCV
face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')#Load the cascade
```

```python
#Upload the input image
img = cv2.imread('obama.jpg')
```

```python
#Perform face detection
faces = face_cascade.detectMultiScale(img, 1.1, 4) #Detect faces
```

```python
#Drawing rectangle around the faces
for (x, y, w, h) in faces:
  cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 0), 2)
```

```python
#Exporting the image file
cv2.imwrite("face_detected.jpg", img)
print('Photo successfully exported!')
```
